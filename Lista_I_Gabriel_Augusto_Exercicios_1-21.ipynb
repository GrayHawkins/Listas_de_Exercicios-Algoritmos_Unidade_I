```python
# 21. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário considerando os descontos de IR (11%), INSS (8%) e Sindicato (5%).

valor_hora = float(input("Quanto você ganha por hora? R$ "))
horas_trabalhadas = float(input("Quantas horas trabalhou no mês? "))

salario_bruto = valor_hora * horas_trabalhadas

ir = salario_bruto * 0.11
inss = salario_bruto * 0.08
sindicato = salario_bruto * 0.05

descontos = ir + inss + sindicato
salario_liquido = salario_bruto - descontos

print("Salário Bruto : R$", salario_bruto)
print("IR (11%)      : R$", ir)
print("INSS (8%)     : R$", inss)
print("Sindicato (5%): R$", sindicato)
print("Salário Líquido: R$", salario_liquido)
```

```python
# 20. Ler dois valores para as variáveis A e B, e efetuar as trocas dos valores de forma que a variável A passe a possuir o valor da variável B e a variável B passe a possuir o valor da variável A.

A = input("Digite o valor de A: ")
B = input("Digite o valor de B: ")

temp = A
A = B
B = temp

print("Novo valor de A:", A)
print("Novo valor de B:", B)
```

```python
# 19. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em MBps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

tamanho = float(input("Digite o tamanho do arquivo (MB): "))
velocidade = float(input("Digite a velocidade da internet (MBps): "))

tempo_segundos = tamanho / velocidade
tempo_minutos = tempo_segundos / 60

print("Tempo aproximado de download:", tempo_minutos, "minutos")
```

```python
# 17. Faça um Programa que peça 2 números inteiros e um número real. Calcule e mostre:
# - o produto do dobro do primeiro com metade do segundo.
# - a soma do triplo do primeiro com o terceiro.
# - o terceiro elevado ao cubo.

num1 = int(input("Digite o primeiro número inteiro: "))
num2 = int(input("Digite o segundo número inteiro: "))
num3 = float(input("Digite o número real: "))

resultado1 = (num1 * 2) * (num2 / 2)
resultado2 = (num1 * 3) + num3
resultado3 = num3 ** 3

print("Produto do dobro do primeiro com metade do segundo:", resultado1)
print("Soma do triplo do primeiro com o terceiro:", resultado2)
print("Terceiro elevado ao cubo:", resultado3)
```

```python
# 16. Escrever um algoritmo que leia o nome de um vendedor, o seu salário fixo e o total de vendas efetuadas por ele no mês. Sabendo que este vendedor ganha 15% de comissão sobre suas vendas, informar o seu nome, o salário fixo e o salário no final do mês.

nome = input("Digite o nome do vendedor: ")
salario_fixo = float(input("Digite o salário fixo: R$ "))
vendas = float(input("Digite o valor total das vendas: R$ "))

comissao = vendas * 0.15
salario_final = salario_fixo + comissao

print("Nome:", nome)
print("Salário fixo: R$", salario_fixo)
print("Salário final: R$", salario_final)
```

```python
# 15. Escreva um programa que leia três números inteiros e positivos (A, B, C) e calcule a seguinte expressão: D = (R + S) / 2, onde R = (A + B)² e S = (B + C)²

A = int(input("Digite o valor de A: "))
B = int(input("Digite o valor de B: "))
C = int(input("Digite o valor de C: "))

R = (A + B) ** 2
S = (B + C) ** 2

D = (R + S) / 2

print("O valor de D é:", D)
```

```python
# 13. Faça um algoritmo que receba o preço de custo de um produto e mostre o valor de venda. O preço de custo receberá um acréscimo de acordo com um percentual informado pelo usuário.

preco_custo = float(input("Digite o preço de custo: R$ "))
percentual = float(input("Digite o percentual de acréscimo (%): "))

valor_venda = preco_custo + (preco_custo * percentual / 100)

print("Valor de venda: R$", valor_venda)
```

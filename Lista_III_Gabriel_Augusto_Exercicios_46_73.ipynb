```python
# 73. Faça um programa para validar o login e a senha.
# Login válido: "kezia"
# Senha válida: "123"
# O usuário possui no máximo 3 tentativas.

tentativas = 0

while tentativas < 3:
    login = input("Login: ")
    senha = input("Senha: ")

    if login == "kezia" and senha == "123":
        print("Login realizado com sucesso!")
        break
    else:
        tentativas += 1
        print("Login ou senha inválidos!")

if tentativas == 3:
    print("Número máximo de tentativas excedido.")
```

```python
# 72. Faça um programa que receba um número
# e verifique se ele é ou não triangular.

numero = int(input("Digite um número: "))

i = 1
triangular = False

while i * (i + 1) * (i + 2) <= numero:
    if i * (i + 1) * (i + 2) == numero:
        triangular = True
        break
    i += 1

if triangular:
    print("O número é triangular.")
else:
    print("O número não é triangular.")
```

```python
# 71. Faça um programa que calcule quantos pontos fez um time
# em um campeonato de futebol.
# Vitória = 3 pontos
# Empate = 1 ponto
# Derrota = 0 ponto
# A leitura termina quando o número de gols do time for negativo.

pontos = 0

while True:
    gols_time = int(input("Gols do time: "))

    if gols_time < 0:
        break

    gols_adversario = int(input("Gols do adversário: "))

    if gols_time > gols_adversario:
        pontos += 3
    elif gols_time == gols_adversario:
        pontos += 1

print("Total de pontos:", pontos)
```

```python
# 70. Faça um programa que leia o código dos itens pedidos
# e as quantidades desejadas. Calcule o valor de cada item
# e o total geral do pedido. O código 0 encerra o pedido.

total = 0

while True:
    print("\nCardápio")
    print("100 - Cachorro Quente - R$ 1.20")
    print("101 - Bauru Simples - R$ 1.30")
    print("102 - Bauru com Ovo - R$ 1.50")
    print("103 - Hambúrguer - R$ 1.20")
    print("104 - Cheeseburguer - R$ 1.30")
    print("105 - Refrigerante - R$ 1.00")
    print("0 - Encerrar pedido")

    codigo = int(input("Código: "))

    if codigo == 0:
        break

    quantidade = int(input("Quantidade: "))

    if codigo == 100:
        preco = 1.20
    elif codigo == 101:
        preco = 1.30
    elif codigo == 102:
        preco = 1.50
    elif codigo == 103:
        preco = 1.20
    elif codigo == 104:
        preco = 1.30
    elif codigo == 105:
        preco = 1.00
    else:
        print("Código inválido!")
        continue

    valor = preco * quantidade
    total += valor

    print("Valor do item: R$", valor)

print("Total do pedido: R$", total)
```

```python
# 69. Faça um programa que leia uma quantidade não determinada de números
# positivos. Calcule a quantidade de números pares e ímpares,
# a média dos valores pares e a média geral dos números lidos.
# O número que encerrará a leitura será 0.

quantidade = 0
soma = 0

pares = 0
soma_pares = 0

impares = 0

while True:
    numero = int(input("Digite um número (0 para encerrar): "))

    if numero == 0:
        break

    quantidade += 1
    soma += numero

    if numero % 2 == 0:
        pares += 1
        soma_pares += numero
    else:
        impares += 1

if quantidade > 0:
    media_geral = soma / quantidade

    if pares > 0:
        media_pares = soma_pares / pares
    else:
        media_pares = 0

    print("Quantidade de pares:", pares)
    print("Quantidade de ímpares:", impares)
    print("Média dos valores pares:", media_pares)
    print("Média geral:", media_geral)
else:
    print("Nenhum número foi informado.")
```

```python
# 68. A prefeitura de uma cidade fez uma pesquisa entre seus habitantes,
# coletando dados sobre o salário e número de filhos.
# Calcule:
# a) Média do salário da população;
# b) Média do número de filhos;
# c) Maior salário;
# d) Percentual de pessoas com salário até R$ 250,00.
# O programa deve encerrar quando for informado um salário negativo.

soma_salarios = 0
soma_filhos = 0
maior_salario = 0
quantidade_pessoas = 0
salarios_ate_250 = 0

while True:
    salario = float(input("Digite o salário (negativo para encerrar): "))

    if salario < 0:
        break

    filhos = int(input("Digite o número de filhos: "))

    soma_salarios += salario
    soma_filhos += filhos
    quantidade_pessoas += 1

    if salario > maior_salario:
        maior_salario = salario

    if salario <= 250:
        salarios_ate_250 += 1

if quantidade_pessoas > 0:
    media_salario = soma_salarios / quantidade_pessoas
    media_filhos = soma_filhos / quantidade_pessoas
    percentual = (salarios_ate_250 / quantidade_pessoas) * 100

    print("Média do salário: R$", media_salario)
    print("Média do número de filhos:", media_filhos)
    print("Maior salário: R$", maior_salario)
    print("Percentual com salário até R$250,00:", percentual, "%")
else:
    print("Nenhum dado foi informado.")
```

```python
# 63. Faça um programa em Python para mostrar os N primeiros termos
# da série de FETUCCINE, sabendo-se que para existir esta série
# serão necessários pelo menos três termos.

n = int(input("Digite a quantidade de termos (mínimo 3): "))

while n < 3:
    n = int(input("Valor inválido! Digite um número maior ou igual a 3: "))

a1 = int(input("Digite o primeiro termo: "))
a2 = int(input("Digite o segundo termo: "))

print("Série de FETUCCINE:")
print(a1)
print(a2)

anterior2 = a1
anterior1 = a2

for i in range(3, n + 1):
    if i % 2 == 0:
        atual = anterior1 + anterior2
    else:
        atual = anterior1 - anterior2

    print(atual)

    anterior2 = anterior1
    anterior1 = atual
```

```python
# 57. Cada espectador de um cinema respondeu a um questionário
# com sua idade e opinião sobre o filme:
# ótimo = 3, bom = 2, regular = 1.
# Receba os dados de 15 espectadores e calcule:
# a) média das idades que responderam ótimo;
# b) quantidade que responderam regular;
# c) porcentagem que respondeu bom.

soma_idades_otimo = 0
qtd_otimo = 0
qtd_regular = 0
qtd_bom = 0

for i in range(1, 16):
    idade = int(input(f"Idade do espectador {i}: "))
    opiniao = int(input("Opinião (3-Ótimo, 2-Bom, 1-Regular): "))

    if opiniao == 3:
        soma_idades_otimo += idade
        qtd_otimo += 1
    elif opiniao == 2:
        qtd_bom += 1
    elif opiniao == 1:
        qtd_regular += 1

if qtd_otimo > 0:
    media_otimo = soma_idades_otimo / qtd_otimo
else:
    media_otimo = 0

porcentagem_bom = (qtd_bom / 15) * 100

print("Média das idades (ótimo):", media_otimo)
print("Quantidade de respostas regular:", qtd_regular)
print("Porcentagem de respostas bom:", porcentagem_bom, "%")
```

```python
# 56. Faça um programa que mostre os n termos da série:
# S = 1/1 + 2/3 + 3/5 + 4/7 + ... + n/m.
# Imprima no final a soma da série.

n = int(input("Digite a quantidade de termos: "))

numerador = 1
denominador = 1
soma = 0

for i in range(n):
    termo = numerador / denominador
    print(f"{numerador}/{denominador} = {termo}")

    soma += termo
    numerador += 1
    denominador += 2

print("Soma da série =", soma)
```

```python
# 55. Faça um programa que recebe o número real x como entrada e
# devolva uma aproximação do arco tangente de x (em radianos)
# através da série:
# arctan(x) = x - x³/3 + x⁵/5 - x⁷/7 + ...
# Considere a aproximação para 50 termos.

x = float(input("Digite o valor de x: "))

arctan = 0
sinal = 1
expoente = 1

for denominador in range(1, 100, 2):
    termo = (x ** expoente) / denominador
    arctan += sinal * termo

    sinal *= -1
    expoente += 2

print("Aproximação de arctan(x) =", arctan)
```
